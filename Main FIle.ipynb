## Importing libraries

```python
import pandas as pd
import numpy as np
import json
import sqlite3
```

## Data input

```python
with open("employee.json", "r+") as f:
    json_data = json.load(f)
json_data    
```

```python
type(json_data)
```

```python
json_data.keys()
```

```python
json_data.values()
```

```python
value = {}
for item in json_data.values():
    value = item
value    
```

## Converting JSON Data into DataFrame

```python
df = pd.DataFrame(value)
df.head(5)
```

```python
df.columns
```

## converting DataFrame columns into lists

```python
id_df = df['ID'].tolist()
job_title_df = df['JobTitle'].tolist()
email_address_df = df['EmailAddress'].tolist()
firstname_lastname_df = df['FirstNameLastName'].tolist()
vaccinated_df = df['vaccinated'].tolist()
```

```python
id_df = [eval(i) for i in id_df]
id_df
```

```python
## setting len of list
n = len(id_df)
```

## SQLITE Setup

```python
connection = sqlite3.connect("data.db")
```

```python
csr = connection.cursor()
```

#### creating new table employee

```python
sql_command_create_emp_table = """ CREATE TABLE employee (
id INTEGER PRIMARY KEY,
JobTitle VARCHAR(100),
EmailAddress VARCHAR(100),
FirstNameLastName VARCHAR(100),
vaccinated BOOLEAN
)"""

csr.execute(sql_command_create_emp_table)
```

#### setting values for table employee

```python
for i in range(n):
    csr.execute("""INSERT INTO employee(id, JobTitle, EmailAddress, FirstNameLastName, vaccinated) VALUES (?,?,?,?,?);""",(id_df[i], job_title_df[i], email_address_df[i], firstname_lastname_df[i], vaccinated_df[i]))
```

#### Fetching Data

```python
csr.execute("SELECT * FROM employee")
ans = csr.fetchall()
for i in ans:
    print(i)
```

```python
## All columns in table "employee"
data = csr.execute('''SELECT * FROM employee''')
for column in data.description:
    print(column[0])
```

#### adding new column 'bonus percent' for giving bonus percent
#### - percent is 15% if person is vaccinated
#### - percent is 5 % if person is not vaccinated



```python
### adding new column

csr.execute(f"""
ALTER TABLE employee 
ADD COLUMN bonus_percent INTEGER
""")
```

```python
### setting data for column - bonus_percent as 15 for people with vaccinated=True
csr.execute("""
UPDATE employee
SET bonus_percent = 15
WHERE vaccinated = 'True'
""")
```

```python
### setting data for column - bonus_percent as 5 for people with vaccinated=False
csr.execute("""
UPDATE employee
SET bonus_percent = 5
WHERE vaccinated = 'False'
""")
```

```python
## fetching data
csr.execute("SELECT * FROM employee")
ans = csr.fetchall()
for i in ans:
    print(i)
```
